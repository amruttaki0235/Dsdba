```python
sorted_rscore= sorted(df['reading_score'])
```

```python
sorted_rscore
```

```python
q1 = np.percentile(sorted_rscore, 25)
```

```python
q3 = np.percentile(sorted_rscore, 75)
```

```python
print(q1,q3)
```

```python
ior=q3-q1
```

```python
ior
```

```python
lb=q1-(1.5*ior)
```

```python
ub=q3+(1.5*ior)
```

```python
print(lb,ub)
```

```python
outlier=[]
for i in df['reading_score']:
    if(i<lb or i>ub):
       outlier.append(i)
outlier
```

```python
ndf=df
ndf
```

```python
outlier_rows = ndf[ndf['reading_score'].isin(outlier)].index
```

```python
df.drop(outlier_rows)
```
