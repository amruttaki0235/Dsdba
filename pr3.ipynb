```python
victim=df['Number of Reported Victims']
```

```python
desc=victim.describe()
desc
```

```python
v_mean = desc['mean']
v_mode = desc.mode()  # Consider extracting a single mode value if needed
v_median = desc.median()
v_std = desc['std']
v_q1 = desc['25%']
v_q3 = desc['75%']
v_variance = v_std ** 2
v_mdev = abs(sum(desc - v_mean))  # Assuming desc contains the data

# Create a dictionary containing the statistics
stats = {
    'v_mean': v_mean,
    'v_mode': v_mode,
    'v_median': v_median,
    'v_std': v_std,
    "v_q1": v_q1,
    "v_q3": v_q3,
    'v_variance': v_variance,
    'v_mdev': v_mdev
}
stats
```

```python
skewness=(3*(v_mean-v_median))/v_std
```

```python
skewness
```

```python
if skewness>0:
    print('right skewness')
else:
    print("left")
```
